# **IMAGE PROCESSING TECHNIQUES**


### Input image

```python
from PIL import Image
img = Image.open("_jelly_fish.jpg")
img.show()
```

## **Performing image processing operations on the given image:**

### 1. Gray Image

```python
gray = img.convert('L')
gray.save('gray.jpg')
gray.show()
```

### 2. Transpose Image

```python
transpose = img.transpose(Image.TRANSPOSE)
transpose.show()
transpose.save('transpose.jpg')
```

### 3. Inverted Image

```python
from PIL import ImageOps
inverted = ImageOps.invert(img)
inverted.show()
inverted.save('inverted.jpg')
```

### 4. Blurred Image

```python
from PIL import ImageFilter
blur = img.filter(ImageFilter.BoxBlur(10))
blur.show()
blur.save('box_blur_radius_10.jpg')
```

### 5. Finding image edges

```python
edges = img.filter(ImageFilter.FIND_EDGES)
edges.show()
edges.save('find_edges.jpg')
```

### 6. Enhancing image edges

```python
enhance = img.filter(ImageFilter.EDGE_ENHANCE)
enhance.show()
enhance.save('enhance.jpg')
```

### 7. Unsharp masking on image

```python
unsharp = img.filter(ImageFilter.UnsharpMask(10))
unsharp.show()
unsharp.save('unsharp.jpg')
```
