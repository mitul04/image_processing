```python
from PIL import ImageFilter
img1.filter(ImageFilter.BLUR).save("Blur.jpg") # Soft Blur
img1.filter(ImageFilter.BoxBlur(25)).save("BoxBlur(20).jpg")
img1.filter(ImageFilter.GaussianBlur(25)).save("GaussBlur(20).jpg")
```

```python
img2 = Image.open('_jelly_fish.jpg')
img2.filter(ImageFilter.CONTOUR).save("Contour.jpg")
img2.filter(ImageFilter.DETAIL).save("Detail.jpg")
img2.filter(ImageFilter.EDGE_ENHANCE).save("EdgeEnhance.jpg")
img2.filter(ImageFilter.EDGE_ENHANCE_MORE).save("EdgeEnhanceMore.jpg")
img2.filter(ImageFilter.EMBOSS).save("Emboss.jpg")
img2.filter(ImageFilter.SMOOTH).save("Smooth.jpg")
img2.filter(ImageFilter.SHARPEN).save("Sharpen.jpg")
img2.filter(ImageFilter.UnsharpMask(5)).save("UnsharpMask(5).jpg")
img2.filter(ImageFilter.FIND_EDGES).save("FindEdges.jpg")
```

```python
from PIL import ImageDraw
img3 = Image.open('_Halloween_2023.jpeg')
w, h = img3.size

draw = ImageDraw.Draw(img3)

draw.rectangle((w // 4, h // 4, 3 * w // 4, 3 * h // 4), outline="red", width=5)
cropped1 = img3.crop((w // 4, h // 4, 3 * w // 4, 3 * h // 4))
cropped1.save("cropped1_red.jpg")

# Crop 800 pixels from left, 650 pixels from top, 20% pixels from right, and 250 pixels from bottom
draw.rectangle((800, 650, w - 0.2*w, h - 250), outline="yellow", width=5)
cropped2 = img3.crop((800, 650, w - 0.2*w, h - 250))
cropped2.save("cropped2_yellow.jpg")

img3.save("draw_rectangle.jpg")
```

```python
from PIL import ImageOps
img4 = Image.open('_Halloween_2023.jpeg')
inverted = ImageOps.invert(img4)
inverted.save("inverted.jpg")
```

```python
img5 = Image.open('_text.png')
flipped1 = img5.transpose(Image.FLIP_LEFT_RIGHT)
flipped2 = img5.transpose(Image.FLIP_TOP_BOTTOM)
flipped3 = img5.transpose(Image.TRANSPOSE)
flipped4 = img5.transpose(Image.Transpose.ROTATE_90) # anti-clockwise
flipped5 = img5.transpose(Image.Transpose.ROTATE_180) # same as LEFT_RIGHT + TOP_BOTTOM
flipped6 = img5.transpose(Image.TRANSVERSE) # 180 rotation of TRANSPOSE

flipped1.save("text_flipped1.png") # flipped1.convert("RGB").save("text_flipped1.jpg")
flipped2.save("text_flipped2.png")
flipped3.save("text_flipped3.png")
flipped4.save("text_flipped4.png")
flipped5.save("text_flipped5.png")
flipped6.save("text_flipped6.png")
```

```python
img6 = Image.open('_dine.jpg')
```

```python
from PIL import ImageOps
border_img = ImageOps.expand(img6, border=(50, 100, 200, 300), fill=(0, 0, 255)).save("bordered.jpg")
```

```python
from PIL import ImageEnhance
enhance = ImageEnhance.Brightness(img6)
enhance.enhance(0.5).save("brightness_0.5.jpg")
enhance.enhance(2.5).save("brightness_2.5.jpg")

enhance = ImageEnhance.Contrast(img6)
enhance.enhance(0.5).save("contrast_0.5.jpg")
enhance.enhance(1.5).save("contrast_1.5.jpg")

enhance = ImageEnhance.Sharpness(img6)
enhance.enhance(0.5).save("sharpness_0.5.jpg")
enhance.enhance(3).save("sharpness_3.jpg")

enhance = ImageEnhance.Color(img6)
enhance.enhance(0.8).save("color_0.8.jpg")
enhance.enhance(2.0).save("color_2.0.jpg")
```

```python
import os

if not os.path.exists('2'):
    os.mkdir('2')

for f in os.listdir('.'):
    f = f.lower()
    if (f.endswith('.jpg') or f.endswith('.png')) and not f.startswith('_'):
        file = Image.open(f)
        f_name, f_ext = os.path.splitext(f)
        file.save(f"2/{f_name} (copied).png")
        os.remove(f)
    
```
